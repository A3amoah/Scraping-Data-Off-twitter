```python
import pandas as pd   # import pandas library for working with dataframes
import snscrape.modules.twitter as sntwitter # import sntwitter library for accessing Twitter data
from tqdm import tqdm             # import tqdm library for showing progress bars

user_handle = "BarackObama"        # specify the Twitter handle of the user to retrieve tweets from
query = f"from:{user_handle}"      # construct the query string to search for tweets from the user
scraper = sntwitter.TwitterSearchScraper(query)  # create a TwitterSearchScraper object to retrieve tweets
tweets = []                        # initialize an empty list to store the tweet data
n_tweets = 1000                    # specify the maximum number of tweets to retrieve

# loop over the tweets returned by the scraper, using tqdm to display a progress bar
for i, tweet in tqdm(enumerate(scraper.get_items()), total=n_tweets):
    # extract relevant data from the tweet object and store it as a list
    data = [
        tweet.date,
        tweet.id, 
        tweet.content,
        tweet.user.username,
        tweet.likeCount,
        tweet.retweetCount,
    ]
    # append the list of tweet data to the tweets list
    tweets.append(data)
    # break out of the loop if we have retrieved the desired number of tweets
    if i >= n_tweets - 1:
        break

# convert the list of tweet data into a pandas dataframe
tweet_df = pd.DataFrame(
    tweets, columns=["date", "id", "content", "username", "like_count", "retweet_count"]
)

# save the dataframe to a CSV file, using the user's handle in the filename
tweet_df.to_csv(f"{user_handle}-tweets.csv", index=False)
```
